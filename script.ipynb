Plotando gráficos

```python
import seaborn as sns
```

```python
ax = sns.distplot(df['Valor de Venda'])

ax.figure.set_size_inches(12, 6)
ax.set_title('Distribuição de Frequência - Valor de Venda', fontsize=20)
ax.set_ylabel('Quantidade', fontsize=16)
ax
```

```python
df['Valor de Venda'].describe()
```

```python
df.columns
```

```python
ax = sns.boxplot(y='Valor de Venda', x='Estado - Sigla', data=df, orient='v', hue='Estado - Sigla')

ax.figure.set_size_inches(12, 6)
ax.set_title('Distribuição Valor de Venda', fontsize=20)
ax.set_ylabel('Valor', fontsize=16)
ax.set_xlabel('Estados', fontsize=16);
```
